```python
import builtins
dir(builtins)
```

```python
def square():
    """return square of value"""
    def add():
        """add two local variable """
        x = 2
        y = 3 
        z = x + y
        return z
    return add()**2
print (square())
```

```python
def f(a, b=1, c=2):
    y = a+b+c
    return y
print(f(5))
print(f(5,4))
print(f(5,4,3))
```

```python
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)

def f(**kwargs):
    """ print key and value of dictionary """
    for key, value in kwargs.items():
        print(key, " ", value)
f(country="spain", capital="madrid")
```
